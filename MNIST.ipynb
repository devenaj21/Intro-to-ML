```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import decomposition

pca = decomposition.PCA()

data = pd.read_pickle('/content/drive/MyDrive/data/data/mnist_small.pkl')
images = np.array(data['X'])
labels = np.array(data['Y'])
pca.n_components = 2
pca_data = pca.fit_transform(images)
```

```python
import seaborn as sns
# plt.scatter(pca_data[:,0],pca_data[:,1],color=1)
pca_data = np.c_[labels, pca_data]
cols = ['label', 'Dim 1', 'Dim 2']
pca_data = pd.DataFrame(pca_data, columns=cols)
print(pca_data.shape)
```

```python
sns.lmplot(x='Dim 1', y ='Dim 2', hue ='label', data=pca_data)
plt.show()
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2)
tsne_data = model.fit_transform(images)
tsne_data = np.c_[labels, tsne_data]
cols = ['label', 'Dim 1', 'Dim 2']
tsne_data = pd.DataFrame(tsne_data, columns=cols)
print(pca_data.shape)
```

```python
sns.lmplot(x='Dim 1', y ='Dim 2', hue ='label', data=tsne_data)
plt.show()
```
